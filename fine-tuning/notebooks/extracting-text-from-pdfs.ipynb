## Extracting text from pdf file 
**Library used**: `pymupdf`

```python
import fitz # pymupdf lib.
```

```python
FILE_PATH=r"E:\genai\fine-tuning\dataset\updated_file.pdf"

file_content = []
with fitz.open(FILE_PATH) as pdf_file:
    for page in pdf_file:
        file_content.append(page.get_text())
print("[+] File reading done")
```

### cleaning the file content.

```python
# remove the smaller sentences, first of all split the whole text on the basis of `.` and remove sentence whihc are smaller than 15 words.

def preprocess_data(sentences: str):
    lower_case_sentences = sentences.lower()
    return lower_case_sentences.split(".")

filtered_content = []
for page_content in file_content:
    clean_data = preprocess_data(page_content)
    for sentence in clean_data:
        
        clean_sentence = sentence.strip().replace("-\n", "").replace("\n", " ")
        if len(clean_sentence.split()) > 15:
            filtered_content.append(clean_sentence)
```
